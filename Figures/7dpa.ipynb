```python
import pandas as pd
import numpy as np
import os
import scanpy as sc
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import cmocean.cm as cmo
import cmocean
```

```python
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = "Arial"
```

```python
adata = sc.read('../../Data/Visium/Visium_7dpa.h5ad')
```

```python
Edges_3D = adata.uns['Edges_3D']
color_map = dict(zip(adata.obs['STAGATE'].cat.categories, adata.uns['STAGATE_colors']))
```

# Without edges

```python
#####################
fig = plt.figure(figsize=[5,6])
ax = fig.add_subplot(111, projection='3d') 
ax.set_box_aspect(aspect = (2.2,1,1))

ax.set_xlim([0, 11000])
ax.set_ylim([0, 8000])
ax.set_zlim([0, 500])
plt.axis('off')

ax.set_title("7 dpa")


for it, label_it in enumerate(np.unique(adata.obs['STAGATE'])):
    ad = adata[adata.obs.loc[adata.obs['STAGATE']==label_it,].index,]
    ax.scatter(ad.obs['X_3D'], ad.obs['Y_3D'], ad.obs['Z_3D']*30, alpha=0.8, marker='o',label=label_it,s=6, linewidths=0, c=color_map[label_it])
    
# Plot edges
#ax.scatter(Edges_3D['Aligned_X'], Edges_3D['Aligned_Y'], Edges_3D['Section_id']*30, c='grey', marker='.', s=1, alpha=0.1)

plt.legend(bbox_to_anchor =[1, 1])

# View setting
ax.view_init(elev=30, azim=70)

plt.savefig('Images/7dpa/STAGATE.pdf')
```

```python
for show_label in np.unique(adata.obs['STAGATE']):
    #####################
    fig = plt.figure(figsize=[5,6])
    ax = fig.add_subplot(111, projection='3d') 
    ax.set_box_aspect(aspect = (2.2,1,1))

    ax.set_xlim([0, 11000])
    ax.set_ylim([0, 8000])
    ax.set_zlim([0, 500])
    plt.axis('off')

    ax.set_title(show_label)

    for it, label_it in enumerate(np.unique(adata.obs['STAGATE'])):
        ad = adata[adata.obs.loc[adata.obs['STAGATE']==label_it,].index,]
        if label_it==show_label:
            ax.scatter(ad.obs['X_3D'], ad.obs['Y_3D'], ad.obs['Z_3D']*30, alpha=0.8, marker='o',label=label_it,
                       s=6, linewidths=0, c=color_map[label_it])
        else:
            ax.scatter(ad.obs['X_3D'], ad.obs['Y_3D'], ad.obs['Z_3D']*30, alpha=0.1, marker='o',label=label_it,
                       s=6, linewidths=0, c='grey')
    # Plot edges
    #ax.scatter(Edges_3D['Aligned_X'], Edges_3D['Aligned_Y'], Edges_3D['Section_id']*30, c='grey', marker='.', s=1, alpha=0.1)

    #plt.legend(bbox_to_anchor =[1, 1])

    # View setting
    ax.view_init(elev=30, azim=70)
    
    plt.savefig('Images/7dpa/STAGATE_'+show_label+'.pdf')
```
