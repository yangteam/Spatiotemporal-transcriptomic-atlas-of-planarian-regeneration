```python
import pandas as pd
import numpy as np
import scanpy as sc
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = "Arial"
```

```python
adata = sc.read('../../Data/adata_scRNA_Annotated.h5ad')
```

```python
sc.set_figure_params(figsize=[10, 10])
sc.pl.umap(adata, color='Annotation', legend_loc='on data', s=12,#groups=['neurons_UNK_5'],
           add_outline=True, legend_fontsize=12, show=False, legend_fontoutline=2)

plt.savefig('scRNA_UMAP.png', bbox_inches='tight')
```
