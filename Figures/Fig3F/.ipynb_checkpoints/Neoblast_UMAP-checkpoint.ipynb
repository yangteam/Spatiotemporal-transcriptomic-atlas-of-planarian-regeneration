```python
import pandas as pd
import numpy as np
import scanpy as sc
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = "Arial"
```

```python
adata = sc.read('../../Data/adata_Neoblast.h5ad')
```

```python
sc.set_figure_params(figsize=[3, 3])
sc.pl.umap(adata, color='SMED30031920', show=False,legend_fontsize=7, legend_fontoutline=1,
           add_outline=False, legend_loc='on data', title='Louvain (resolution=1.5)')#, save='Louvain_1p5.png')
plt.xlabel('')
plt.ylabel('')
plt.title('')
#plt.savefig('Neoblast_Annotation.png', bbox_inches='tight')
```
