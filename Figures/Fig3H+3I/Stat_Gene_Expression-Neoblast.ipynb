```python
import pandas as pd
import numpy as np
import scanpy as sc
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
sc.set_figure_params(dpi=200, dpi_save=300)
```

```python
adata_Total = sc.read('../../Data/adata_scRNA_Annotated.h5ad')
sc.pp.normalize_total(adata_Total, target_sum=1e4)
sc.pp.log1p(adata_Total)
```

```python
adata_Neoblast = sc.read('../../Data/adata_Neoblast.h5ad')
```

```python
adata = adata_Total[adata_Neoblast.obs_names,]
```

```python
adata.obsm['X_umap'] = pd.DataFrame(adata_Neoblast.obsm['X_umap'], index=adata_Neoblast.obs_names).loc[adata.obs_names,].values
adata.obs['clusters'] = adata_Neoblast.obs.loc[adata.obs_names, 'clusters']
```

```python
sc.set_figure_params(figsize=[3, 3])
sc.pl.umap(adata, color='clusters', show=False, s=5,legend_fontsize=7, legend_fontoutline=1,colorbar_loc=None,vmax='p99.5',
           add_outline=False, legend_loc='on data')
```

```python
marker_list_dict = {}
marker_list_dict['C19'] = ['SMED30031920', 'SMED30003769', 'SMED30014371', 'SMED30003403']

marker_list_dict['C7'] = ['SMED30022580', 'SMED30002753', 'SMED30034500', 'SMED30030310', 'SMED30034477', 'SMED30034470']

marker_list_dict['aa'] = ['SMED30025067', 'SMED30005488', 'SMED30025384']

marker_list_dict['novel'] = ['SMED30022043', 'SMED30012861']
```

```python
symbol_trans_dict = {}
symbol_trans_dict['SMED30031920'] = 'cml13'
symbol_trans_dict['SMED30003769'] = 'capns1a'
symbol_trans_dict['SMED30014371'] = 'smed14371'
symbol_trans_dict['SMED30003403'] = 'smed3403'

symbol_trans_dict['SMED30022580'] = 'runt1'
symbol_trans_dict['SMED30002753'] = 'X1.A5.1'
symbol_trans_dict['SMED30034500'] = 'egr4'
symbol_trans_dict['SMED30030310'] = 'smed30310'
symbol_trans_dict['SMED30034477'] = 'tob2'
symbol_trans_dict['SMED30034470'] = 'pdzrn4'

symbol_trans_dict['SMED30025067'] = 'hspa5'
symbol_trans_dict['SMED30025384'] = 'smed25384'

symbol_trans_dict['SMED30022043'] = 'osr2'
symbol_trans_dict['SMED30012861'] = '2dbd'
```

```python
heatmap_df = pd.DataFrame(index=list(symbol_trans_dict.keys()))
for it in np.unique(adata.obs['clusters']):
    subadata = adata[adata.obs['clusters'] == it]
    sub_df = pd.DataFrame(subadata.X.toarray(), columns=subadata.var_names)
    sub_df = sub_df.loc[:, list(symbol_trans_dict.keys())]
    heatmap_df[str(it)] = np.mean(sub_df)
```

```python
heatmap_df_symbol = heatmap_df.copy()
heatmap_df_symbol.index = [symbol_trans_dict[x] for x in heatmap_df_symbol.index]
```

```python
heatmap_df.to_csv('Neoblast_Markers_Expression.txt', sep='\t')
```

```python
heatmap_df_symbol.to_csv('Neoblast_Markers_Expression_symbol.txt', sep='\t')
```

```python
heatmap_df = heatmap_df.T
heatmap_df = heatmap_df - np.min(heatmap_df)
heatmap_df = heatmap_df/np.max(heatmap_df)
heatmap_df = heatmap_df.T
```

```python
heatmap_df_symbol = heatmap_df.copy()
heatmap_df_symbol.index = [symbol_trans_dict[x] for x in heatmap_df_symbol.index]
```
