```python
import pandas as pd
import numpy as np
import scanpy as sc
import os
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.utils import shuffle
```

```python
plt.rcParams['font.family'] = "sans-serif"
plt.rcParams['font.sans-serif'] = "Arial"
```

```python
adata = sc.read('../../Data/adata_scRNA_Annotated.h5ad')
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
```

```python
sc.set_figure_params(figsize=[3, 3])
sc.pl.umap(adata, color='SMED30008639', s=4, vmax='p99.5', #groups=['neurons_UNK_5'],
           add_outline=True, legend_fontsize=12, colorbar_loc=None,show=False,
           legend_fontoutline=2, title='Louvain (resolution=1.5)')#, save='Louvain_1p5.png')
plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.savefig('MYOD.tif', bbox_inches='tight')
plt.savefig('MYOD.pdf', bbox_inches='tight')
```
